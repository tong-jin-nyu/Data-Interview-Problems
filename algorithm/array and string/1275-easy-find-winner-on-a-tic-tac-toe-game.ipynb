**Tic-tac-toe** is played by two players `A` and `B` on a `3 x 3` grid. The rules of Tic-Tac-Toe 
are:

- Players take turns placing characters into empty squares `' '`.
- The first player `A` always places `'X'` characters, while the second player `B` always places `'O'` characters.
- `'X'` and `'O'` characters are always placed into empty squares, never on filled ones.
- The game ends when there are three of the same (non-empty) character filling any row, column, or diagonal.
- The game also ends if all squares are non-empty.
- No more moves can be played if the game is over.

Given a 2D integer array `moves` where `moves[i] = [rowi, coli]` indicates that the `ith` move 
will be played on `grid[rowi][coli]`. return the winner of the game if it exists (`A` or `B`). 
In case the game ends in a draw return `"Draw"`. If there are still movements to play return 
`"Pending"`.

You can assume that `moves` is valid (i.e., it follows the rules of Tic-Tac-Toe), 
the grid is initially empty, and `A` will play first.

**Example 1:**

![example-1](https://assets.leetcode.com/uploads/2021/09/22/xo1-grid.jpg)

``` text
Input: moves = [[0,0],[2,0],[1,1],[2,1],[2,2]]
Output: "A"
Explanation: A wins, they always play first.

```

**Example 2:**

![example-2](https://assets.leetcode.com/uploads/2021/09/22/xo2-grid.jpg)

``` text
Input: moves = [[0,0],[1,1],[0,1],[0,2],[1,0],[2,0]]
Output: "B"
Explanation: B wins.
```

**Example 3:**

![example-3](https://assets.leetcode.com/uploads/2021/09/22/xo3-grid.jpg)

``` text
Input: moves = [[0,0],[1,1],[2,0],[1,0],[1,2],[2,1],[0,1],[0,2],[2,2]]
Output: "Draw"
Explanation: The game ends in a draw since there are no moves to make.
```

**Constraints:**

- `1 <= moves.length <= 9`
- `moves[i].length == 2`
- `0 <= rowi, coli <= 2`
- There are no repeated elements on `moves`.
- `moves` follow the rules of tic tac toe.

---


# Java

``` java
class Solution {
    public String tictactoe(int[][] moves) {
        int n = moves.length;
        // An array of 8 counters:
        // counts[0-2]: first three counters count on rows.
        // counts[3-5]: counters 3 to 5 count on columns.
        // counts[6]: counter 6 count on the primary diagonal.
        // counts[7]: counter 7 count on the secondary diagonal.
        int[] counts = new int[8];

        // Use a reverse for loop to track the person who took the last move.
        for (int i = n - 1; i >= 0; i -= 2) {
            counts[moves[i][0]]++;
            counts[moves[i][1] + 3]++;
            if (moves[i][0] == moves[i][1])
                counts[6]++;
            if (moves[i][0] + moves[i][1] == 2)
                counts[7]++;
            // If any of the counter reaches 3, 
            // return winner "A" if the length of moves is odd;
            // return winner "B" otherwise.
            if (counts[moves[i][0]] == 3 || counts[moves[i][1] + 3] == 3 ||
                    counts[6] == 3 || counts[7] == 3)
                return n % 2 == 0 ? "B" : "A";
        }
        // Return "Pending" if there's no winner after the for loop.
        if (n < 9)
            return "Pending";
        return "Draw";
    }
}
```
